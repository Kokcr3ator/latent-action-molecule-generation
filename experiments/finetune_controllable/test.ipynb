```python
import numpy as np

#generate random transition matrix 5x5
def generate_transition_matrix(size=5):
    matrix = np.random.rand(size, size)
    matrix = matrix / matrix.sum(axis=1, keepdims=True)
    return matrix
#calculate stationary distribution
def calculate_stationary_distribution(transition_matrix):
    eigvals, eigvecs = np.linalg.eig(transition_matrix.T)
    stationary_dist = np.real(eigvecs[:, np.isclose(eigvals, 1)])
    stationary_dist = stationary_dist / stationary_dist.sum()
    return stationary_dist.flatten()

P = generate_transition_matrix()
phi = calculate_stationary_distribution(P)
```

```python
Phi = np.diag(phi)
inv_phi = np.diag(1/phi)
L = Phi - (Phi @ P + P.T @ Phi)/2
Lprime = inv_phi@L@Phi
r = np.random.rand(5)
gamma = 0.99
v = np.linalg.inv(np.eye(5) - gamma*P) @ r

vals1,eig1 = np.linalg.eig(L)
vals2,eig2 = np.linalg.eig(Lprime)

k = 3
idx1 = np.argsort(vals1)[:k]
idx2 = np.argsort(vals2)[:k]
U1 = eig1[:,idx1]
U2 = eig2[:,idx2]
#noramlize U1 and U2
for i in range(k):
    U1[:,i] = U1[:,i]/np.linalg.norm(U1[:,i])
    U2[:,i] = U2[:,i]/np.linalg.norm(U2[:,i])

v1 = U1 @np.linalg.inv(U1.T @ U1)@U1.T @ v
v2 = U2 @np.linalg.inv(U2.T @ U2)@U2.T @ v

```
